```python
import pandas as pd 
import matplotlib.pyplot as plt 
import glob
```

```python
files = glob.glob('data\\2024-10-25\\MAC_90m_30nm\\data\\VAC*')
files
```

```python
for file in files:
    df = pd.read_csv(file)
    up = df.where(df['Direction'] == 1).dropna().sort_values('Voltage')
    down = df.where(df['Direction'] == -1).dropna().sort_values('Voltage')
    plt.figure(figsize=(10, 6), dpi=500)
    plt.title(file)
    plt.scatter(up.Voltage, up.Current, label='up')
    plt.scatter(down.Voltage, down.Current, label='down')
    plt.legend()
    plt.show()
```
